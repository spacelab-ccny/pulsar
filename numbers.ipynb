```python
from prg import *
import torch

key = b"E" * 64
seed = b"seed"

unif = Uniform(Prg(key, seed))
```

```python
x = unif.generate_float32((256,256))
x
```

```python
from matplotlib import pyplot as plt

plt.hist(x.flatten())
```

```python
y = torch.distributions.Normal(0, 1).icdf(torch.tensor(x))
```

```python
plt.hist(y.flatten())
```

```python
z = Normal(Prg(key, seed))
plt.hist(z.generate((256,256)).flatten())
```
